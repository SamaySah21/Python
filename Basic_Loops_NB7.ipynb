```python
num = [1,2,3,4]
```

```python
# basic for loop
for val in num:
    print(val)
```

```python
# break : stop loop iteration
for val in num:
    if val == 3:
        break
    print(val)
```

```python
# continue : skip current iteration and moves to next iteration
for val in num:
    if val == 2:
        continue
    print(val)
```

```python
# range(0,10) : starts from 0 and ends at 9 ---(0,9)
for val in range(0,10):
    print(val)
```

```python
## WHILE Loop:
## Basic while loop
a = 0
while a<10:
    print(a)
    a += 1
print("---- break continue works same as for")
```

Note : For loop runs atleast once but while loop will run only when statement is true ..
