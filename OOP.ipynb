```python
## Using __init__ ( same as constructor in PHP)
## init function is called when object is initialized

class Computer:

    def __init__(self,cpu,ram):
        self.cpu = cpu
        self.ram = ram

    def config(self):
        print("Config is ",self.cpu, self.ram)

com1 = Computer('i5',16)                                        ## object initialization
com2 = Computer('Ryzen 3',8)                                    ## object initialization


com1.config()
com2.config()
```

```python
## Constructor Self and Comparing Object

class Computer:

    def __init__(self):
        self.name = "Navin"
        self.age = 28


    def compare(self,other):
        if self.age == other.age:
            return True
        else:
            return False


c1 = Computer()
c1.age = 30          ## comment/uncomment it for different condition
c2 = Computer()


if c1.compare(c2):              ## c1 is self, c2 is other
    print("They are same")
else:
    print("They are different")

# c1.update()

print(c1.age)
print(c2.age)
```

```python
## Types of Variable:
## 2 types : (i) Instance Variable (ii) Class Variable
class Car:

    wheels = 4                       ## class variable

    def __init__(self):
        self.mil = 10                ## instance variable
        self.com = "BMW"


c1 = Car()
c2 = Car()

c1.mil = 8                         ## calling instance variable

Car.wheels = 5                     ## calling class variable


## Also class variable can be called using object

print(c1.com , c1.mil , c1.wheels)
print(c2.com , c2.mil , c2.wheels)
```

```python
## Types of Methods: 3 types:
## (i) Instance Method (ii) Class Method (iii) Static Method

class Student:

    school = 'Telusko'
    
    # constructor method
    def __init__(self,m1,m2,m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    # instance method
    def avg(self):                               
        return (self.m1 + self.m2 + self.m3)/3

    # class method : cls will be passed exactly same as self in instance method
    @classmethod                       # decorator
    def getSchool(cls):
        return cls.school

    # Static Method
    @staticmethod                    # decorator
    def info():
        print("This is Student class.. in abc molude")

    # def get_m1(self):
    #     return self.m1
    #
    # def set_m1(self,value):
    #     self.m1 = value

s1 = Student(34,47,32)
s2 = Student(89,32,12)

print(s1.avg())
print(s2.getSchool())

Student.info()                       ## no need to create object
```

```python
## Inheritance :
class A:
    def feature1(self):
        print("Feature 1 working")

    def feature2(self):
        print("Feature 2 working")

# Single Level Inheritance as it is inheriting only one class i.e "A"
class B(A):
    def feature3(self):
        print("Feature 3 working")

    def feature4(self):
        print("Feature 4 working")

# Multiple Level Inheritance as it is inheriting only 2 class i.e. first B then B aleardy above extended A. So, it got 
# all features of A and B. 
class C(B):
    def feature5(self):
        print("Feature 5 working")

c1 = C()

c1.feature1()
c1.feature2()
c1.feature3()
c1.feature4()
c1.feature5()
```

```python
## Multiple Inhertiance
class A:
    def feature1(self):
        print("first class")

class B:
    def feature2(self):
        print("second class")

class C:
    def feature3(self):
        print("third class")

class D(A,C):                                # inheriting first and third class 
    def feature4(self):
        print("fourth class")

d1 = D()
d1.feature4()
d1.feature1()
d1.feature3()
```

```python
## Dealing with constructor with multiple class
## Important points : 
## (i) If no constructor is present in child class then automatically parent class constructor is called
## (ii) To call parent class constructor explicity , use  "super().__init__()"
## (iii) To call parent class method .. use "super.methodname()"
## (iv)  While using super.methodname() , if there are 2 parent class, then first it will check for left class,
##     if method is not found then it will move to next class. For eg : A(B,C) : first check in B then in C if method is not 
##   found in B.

class A:

    def __init__(self):
        print("in A Init")

    def feature1(self):
        print("Feature 1-A working")

    def feature2(self):
        print("Feature 2 working")

class B:

    def __init__(self):
        # super().__init__()                    ## to call parent class constructor
        print("in B Init")

    def feature3(self):
        print("Feature 1-B working")
    
    def feature2(self):
        print("Feature 2-B working")

    def feature4(self):
        print("Feature 4 working")

class C(B,A):

    def __init__(self):
        super().__init__()                        ## A class constructor will call as first A class will call then B
        print("in C init")


    def feat(self):
        super().feature2()              


a1 = C()
a1.feat()
```
