```python
import pandas as pd
import numpy as np
```

```python
# create series
Series1 = pd.Series([1,2,3,4])
Series1
```

```python
# series with custom index labels
Series1 = pd.Series([1,2,3,4] , index = ["first", "second", "third", "fourth"])
Series1
```

```python
Series2 = pd.Series([5,6,7,8] , index = ["first", "fifth", "sixth", "fourth"])
Series2
```

```python
# Adding 2 series : 2 labels with same name will get add ..if labels are not same then it will give NaN
print(Series1  + Series2)
```

```python
## DataFrame: Concatinating 2 series

column_wse_concat_df = pd.concat([Series1, Series2], axis = 1)
column_wse_concat_df
```

```python
row_wse_concat = pd.concat([Series1, Series2], axis=0)
row_wse_concat
```

```python
## create dataframe using numpy array
df1 = pd.DataFrame(np.random.rand(5,4))
df1
```

```python
df1 = pd.DataFrame(np.random.rand(5,4), index = ["index1","index2","index3","index4","index5"], 
                   columns = ["col1","col2","col3","col4"])
df1
```

```python
## Access columns

print(df1["col1"])                      # single column

print("")

print(df1[["col1", "col2"]])            # Multiple columns
```

```python
## Acces rows by indexname 

print(df1.loc["index3"])             ## using loc i.e. using index name

print("")

print(df1.iloc[2])                  ## using iloc  i.e using index number
```

```python
df1
```

```python
## Access row : 0,2,3 and Access columns : 1,3

df1.iloc[[0,2, 3], [1,3]]
```

```python
## We can do logical indexing in dataframe.

df1 < 0.2
```

```python
## when we subset dataframe , it gives NaN where value is False.

df1[df1 < 0.2]
```

```python
## Add column tp dataframe
## It will autocheck ,, if colname is already exist then it dont create that column

df1["col5"] = np.random.randn(5,1)
df1
```

```python
## to reset index name from dataframe
df1.reset_index()
```

```python
## Above only return with default index , but if we want to make this chgange permanent .. we can use inplace = True 
df1

# df1.reset_index(inplace = True)
```
