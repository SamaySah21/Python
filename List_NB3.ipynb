### LIST
#### Combination of heterogeneous elements.
#### List is a collection which is ordered and changeable i.e list is mutable .
#### List Allows duplicate members .

```python
#Creating list
courses = ["History", "Civics", "Geography", "Chemistry"]
print(courses)
```

```python
## Length of list
print(len(courses))
```

```python
# Access indivual by index from start (starts from 0)
print(courses[0])

#Access last elememnt
last_element = len(courses)-1
print(courses[last_element])

# Last element can be access by using index from back (starts from -1)
print(courses[-1])
```

```python
# Access each by index using loop
for i in range(0, len(courses)):
    print(courses[i])
```

```python
# If we try to access any index which is not present , than it will give IndexedError
print(courses[6])
```

```python
## SLICING
## Access range of value from list i.e will give values from 0 to index 2
print(courses[0:3])

## Access all value after certain index
print(courses[1:])
```

```python
# Add item to end of list (append)
courses.append("art")
print(courses)
```

```python
# add item at certain position of list(insert)
courses.insert(1, "biology")
print(courses)
```

```python
# to add multiple values in list (extend)
courses2 = ["hindi", "english", "maths"]
courses = ['History', 'biology', 'biology', 'Civics', 'Geography', 'Chemistry', 'art', 'art']
# see how it works with insert first
courses.insert(0, courses2)
print(courses)
print("It added whole list inside list")
print(courses[0])


## Now lets do it with extend i.e it add multiple values at end of list
print("")
courses = ['History', 'biology', 'biology', 'Civics', 'Geography', 'Chemistry', 'art', 'art']
courses.extend(courses2)
print(courses)
print(courses[0])
```

```python
# Remove : to remove by its value
courses.remove("art")
print(courses)
```

```python
# Pop : to remove last elemt from list 
# It also return the element which it removed
popped_element = courses.pop()
print(popped_element)
print(courses)
```

```python
# reverse the list
courses.reverse()
print(courses)
```

```python
# Sorting

# Ascending order
courses.sort()
print(courses)

# Descending order
courses.sort(reverse = True)
print(courses)
```

```python
# To get sorted list also without disturb original list .. use sorted()
print(courses) # original list
sorted_list = sorted(courses)
print("")
print(sorted_list)  # sorted list
print("")
print(courses) # original list
```

```python
#getting min and max values
num_list = [2,65,9,0,7,18]
print(min(num_list))
print(max(num_list))
```

```python
# Sum of list
print(sum(num_list))
```

```python
## Get index by value
print(courses.index('art'))
```

```python
## ValueError : if we find index of value which is not present
print(courses.index("botany"))
```

```python
# check using "in" whether element is present in list or not
print("maths" in courses)
print("art" in courses)
```

```python
# enumerate: return index and values
for index, value in enumerate(courses):
    print("Index is {} and Value is {}".format(index, value))
```

```python
# Join : convert list to string (like implode)
join_str = " and ".join(courses)
print(join_str)
```
