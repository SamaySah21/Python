### There are 6 ways to create an array using numpy:
#### array()
#### linspace()
#### logspace()
#### arange()
#### zeros()
#### ones()

```python
## Using array: 
# we can create numpy arrays by converting lists
# this is a vector
narr1 = np.array([1,2,3,4])
print(narr1)

# we can create matrices by converting lists of lists
mat = np.array([[1,2,1],[4,5,9],[1,8,9]])
print('')
print(mat)
print('')
print(mat.T)       # Transpose
```

```python
# Using linspace: Divide the range into some parts i.e. here 10 parts.
## If you dont specify parts than it will break into 50 parts which is default.

vec4 = np.linspace(0,5,10)
print(vec4)
print('')
print(vec4.reshape(5,2))
```

```python
## Using arange: Divide the range using steps i.e. here step by 6
## If we dont specify steps, than by default it will take 1 as step.

vec2 = np.arange(0,15)
print(vec2)
print('')
vec3 = np.arange(3,21,6)
print(vec3)
```

```python
## Using Numpy zeros:
mat2 = np.zeros([5,3])
print(mat2)

print("")
np_zeros = np.zeros(5)
print(np_zeros)
```

```python
## Using Numpy Ones:
mat3 = np.ones((3,5))
print('')
print(mat3)

print("")
np_ones = np.ones(5)
print(np_ones)
```

```python
#Using eye(Identy matrix)
mat4 = np.eye(5)
print('')
print(mat4)
```

```python
# we can (+-*/) arrays together if they're the right size
vec5 = np.arange(1,6)
vec6 = np.arange(3,8)

print(vec5)
print("")
print(vec6)
print("")
print(vec5+vec6)
print("")
print(vec5*vec6)
print("")
print(1/vec5)
print("")
print(np.sqrt(vec6))
print("")
```

```python
### Finding unique in numpy array
num_array = np.array([1,2,32,4,5,5,6,7,8,8,8])
print(np.unique(num_array))
```

```python
# we can also use numpy to generate samples of a random variable and form a unifirm distribution
rand_val = np.random.rand(5)
print(rand_val)

print("")

rand_mat = np.random.rand(5,6)
print(rand_mat)
```

```python
# we can also use numpy to generate samples of a random variable and form a normal distribution
normal_rand_val = np.random.randn(5)
print(normal_rand_val)

print("")

normal_rand_mat = np.random.randn(5,6)
print(normal_rand_mat)
print("")
print("Mean is {}".format(np.mean(normal_rand_mat)))
print("")
print("Std is {}".format(np.std(normal_rand_mat)))
```

```python
## Access one value from numpy array

arr_a = np.arange(19)
print(arr_a)
print("")
print(arr_a[8])     ## Access random value by its index
```

```python
# we can access multiple entries at once using :
print(arr_a[3:10])
```

```python
# we can also access multiple non-consecutive entries using np.arange
print(np.arange(0,10,3))
print(arr_a[np.arange(0,15,3)])
```

```python
# what about matrices
print(rand_mat)
print("")
print(rand_mat[1][2])
print("")
print("or do like below")
print(rand_mat[1,2])
```

```python
## Extract submatrix  : [row, column]

print(rand_mat[0:2,1:3])               ## first fetch row 0,1 then fetch column 1 and 2 from it
print("")
print(rand_mat[2, 3:5])                ## first fetch row 2 then fetch column 3 and 4
```

```python
## CHange some value in array
print(rand_mat)

rand_mat[2,2] = 1

print("")
print(rand_mat)
```

```python
print(rand_mat[1,3])   ## fetching exact
print("")
print(rand_mat[1:3])   ## fetching range
```
