#### Pandas support chaining operation

```python
import pandas as pd
df = pd.read_csv("mall_customer.csv")
```

```python
## select query

''' select * from customer'''
df
```

```python
## LIMIT

''' select * from customer limit 7'''

df.head(7)
```

```python
## SELECT ... WHERE... (Single columns)

''' select Annual Income from customer where Age > 20'''

df[df["Age"] > 20] [["Annual Income (k$)"]]
```

```python
## SELECT ... WHERE...( Multiple Columns)

''' select Annual Income, Gender, Age from customer where Age > 20'''

df[df["Age"] > 20] [["Annual Income (k$)", "Gender", "Age"]]
```

```python
## SELECT ... WHERE...( All)

''' select * from customer where Age > 20'''

df[df["Age"] > 20]
```

```python
## Mutiple condition

'''SELECT * FROM customers WHERE Annual_Income_K_Dollars = 16 AND Spending_Score_1_to_100 = 6'''

df[(df["Gender"] == "Male") & (df["Age"] > 20)].head(5)
```

```python
##### AGGREGATE
```

```python
'''SELECT mean(Age), max(Age), min(Age) FROM customers'''

df.agg({
    "Age": ["mean", "sum"],
    "Spending Score (1-100)" : ["max", "mean"]   
})
```

```python
#### DISTINCT

'''SELECT DISTINCT Annual_Income_K_Dollars FROM customers'''

df["Gender"].unique()
```

```python
## ORDER BY (ASCENDING)
'''SELECT * FROM customers WHERE Gender = "Female" ORDER BY Spending Score (1-100)'''

df[df["Gender"] == "Female"].sort_values("Spending Score (1-100)")
```

```python

#SQL Statement (Filtering on Genre called Female & Ordering By Spending_Score_1_to_100 in DESCENDING order)
'''SELECT * FROM customers WHERE Gender = "Female" ORDER BY Spending Score (1-100) DESC'''

df.groupby(['Gender', 'Age']).size().to_frame('Count').reset_index().sort_values('Age', ascending = False)
```

```python
## GROUPBY count
'''SELECT Genre, Age, count(*) FROM customers GROUPBY Genre, Age'''

df.groupby(["Gender", "Age"]).size()
```

```python
#Formatting above
df.groupby(["Gender", "Age"]).size().to_frame("Count").reset_index()
```

```python
'''SELECT Genre, Age, count(*) FROM customers GROUPBY Genre, Age ORDER BY Age DESC'''

df.groupby(["Gender", "Age"]).size().to_frame("Count").reset_index().sort_values("Age", ascending = False)
```

```python
## HAVING
```

```python
## IN STATEMENT
'''SELECT * FROM customers WHERE Age IN (20, 30, 40)'''

#Pandas Equivalant
# customers[customers.Age.isin([20,30,40])]

df[df['Age'].isin([20,30,40])]
```

```python
## NOT IN STATEMENT
'''SELECT * FROM customers WHERE Age IN (20, 30, 40)'''

#Pandas Equivalant
# customers[customers.Age.isin([20,30,40])]

df[~df['Age'].isin([20,30,40])]
```

```python
#SQL Statement (Identify Top 10 records)
'''SELECT * FROM customers ORDER BY Spending_Score_1_to_100 DESC LIMIT 10'''

#Pandas Equivalant
df.nlargest(10, columns='Spending Score (1-100)')
```

```python
## Above can also be done in this way

df.sort_values("Spending Score (1-100)", ascending = False).head(10)
```

```python
#SQL Statement (Identify Next Top 10 records)
'''SELECT * FROM customers ORDER BY Spending_Score_1_to_100 DESC LIMIT 10 OFFSET 10'''

#Pandas Equivalant
df.nlargest(20, columns='Spending Score (1-100)').tail(10)
```

```python

#SQL Statement (Insert a new record in the table)
'''INSERT INTO customers VALUES(401, 'Male', 50, 30, 20) '''

#Pandas Equivalant
df = df.append({'CustomerID':401, 'Gender':'Male', 'Age':50, 'Annual Income (k$)':30, 'Spending Score (1-100)':20}, ignore_index=True)

df
```

```python
#SQL Statement (Update an existing record in the table)
'''UPDATE customers SET Spending_Score_1_to_100 = 7 WHERE Spending_Score_1_to_100 = 6'''

#Pandas Equivalant
df[df['Spending Score (1-100)'] == 81] = 7

df
```

```python
#SQL Statement (Delete an existing record in the table)
'''DELETE FROM customers WHERE Spending_Score_1_to_100 = 7'''

#Pandas Equivalant

df = df.drop(df[df["Age"] == 7].index)

#df[df["Age"] == 19].index            ## gives rows with true conditions
df.iloc[61]
```
