```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,4])
arr1
```

```python
arr2 = arr1
arr2
```

```python
## Addrress for both arrays are same when we are assigning 
print(id(arr1))
print(id(arr2))
```

```python
## Shalow Copy
arr2 = arr1.view()
```

```python
## Addrress for both arrays are different now when we are making shallow copy

print(id(arr1))

print(id(arr2))
```

```python
## But if we change value in one array , it will change in another array as well.

arr2[2] = 100
arr1[3] = 200

print(arr1)
print(arr2)
```

```python
# To avoid above problem we can use deep copy
arr2 = arr1.copy()
```

```python
## Addrress for both arrays are different now when we are making shallow copy 
print(id(arr1))
print(id(arr2))
```

```python
## But if we change value in one array , it will change in another array as well.
arr1 = np.array([1,2,3,4])
```

```python
# To avoid above problem we can use deep copy
arr2 = arr1.copy()
```

```python
## But if we change value in one array , it will change in another array as well.
arr2[2] = 100
arr1[3] = 200
print(arr1)
print(arr2)
```

#### Deep Copy : value and address are independent for both array
#### Shallow Copy :  address are independent but value are dependent.
