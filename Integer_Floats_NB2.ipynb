### Arithmetic Operators:

```python
# Arithmetic Operation :add(+), subtraction(-), multiplication(*), division(/)
# Comparision Operatpr : smaller than(<), greater than(>), not equal to (<> or !=), equal to (==)
# These things we already know
#Lets go one step ahead from these basic things

#Floor Divsion i.e. gives floor value
print(a//2)

#Square
print(4**2)

#Modulus
print(4%3)
```

```python
#very commonly used : round function
# 2 arguments : value, round upto digits
print(round(4.4))
print(round(4.6))
print(round(4.46890))
print(round(4.46890, 2))
```

```python
## Its always good to do cast while doing arithmentic operation on numbers
num1 = "100"
num2 = 200
res = int(num1) + int(num2)
print(res)
```

```python
## to get list of methods and function avaialble
print(dir(res))
```
