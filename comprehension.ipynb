LIST

```python
# generate a list of 100 number
ls = [i for i in range(100)]
print(ls)
```

```python
# generate a list of 100 number which is divisible by 3
ls = [i for i in range(100) if i%3 == 0]
print(ls)
```

```python
# get only those name which starts with "a" and ends with "e"

ls = ["apple", "banana", "agrape", "avocado", "orange", "pineapple"]
valid_name = [val for val in ls 
              if val[0] == "a" and val[-1] == "e"]
print(valid_name)
```

```python
# flatted a multiple list

ls = [[1,2,3], [4,5,6], [7,8,9]]
flattened_ls = [val1 for val in ls for val1 in val]
print(flattened_ls)
```

```python

#categorize number as even or add

ls = ["Even" if i%2 ==0 else "Odd" for i in range(10)]
print(ls)
```

LIST COMPREHENSION WITH FUNCTION

```python
#generate a list which is square of given range vaues and divided by 2
def square(x):
    return x**2

def even(x):
    return x%2 == 0

ls = [square(x) for x in range(20) if even(x)]
print(ls)
```

DICTIONARY

```python
# dictionary comprehension -> generate a dictionary till 100 number and divisible by 3

dict1 = {i:f"item_{i}" for i in range(100) if i%3 == 0}
print(dict1)
```

```python
# create a dict2 which is opposite of d1 in term of key and value
dict2 = {value:key for key,value in dict1.items()}
print(dict2)
```
