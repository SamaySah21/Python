## Uses Of Lamda : Filter Map Reduce

### Filter : filter(function, sequence)
### Parameters:
#### function: function that tests if each element of a sequence true or not.
#### sequence: sequence which needs to be filtered, it can be sets, lists, tuples, or containers of any iterators.

#### Returns: returns an iterator that is already filtered.

```python
nums = [1,2,3,4,5,6,7,8,9]
```

```python
def is_even(n):
    return n%2 ==0

getEven = list(filter(is_even, nums))
print(getEven)
```

```python
## Filter the above list by even number
## 

getEven = list(filter(lambda n: n% 2 ==0, nums))
getEven
```

### map(fun, iter)
### Parameters:
#### function: It is a function to which map passes each element of given iterable.
#### Iterable: It is a iterable which is to be mapped.

#### Returns: Returns a list of the results after applying the given function  to each item of a given iterable (list, tuple etc.)

```python
## task: whatever even number we got from above filter , we need to double it.
```

```python
def do_double(n):
    return n*2

double = list(map(do_double, getEven))
print(double)
```

```python
## using lambda
double = list(map(lambda n:n*2, getEven))
print(double)
```

```python
# Add two lists using map and lambda 
  
numbers1 = [1, 2, 3] 
numbers2 = [4, 5, 6] 

result = list(map(lambda x,y : x+y, numbers1, numbers2))
print(list(result)) 
```

### Reduce:

#### Working : 

#### At first step, first two elements of sequence are picked and the result is obtained.
#### Next step is to apply the same function to the previously attained result and the number just succeeding the second element and #### the result is again stored.
#### This process continues till no more elements are left in the container.
#### The final returned result is returned and printed on console.

```python
## Reduce always take alteast 2 arguments to start up the working
import functools

print(double)

add_all = functools.reduce(lambda a,b: a+b, double)
add_all
```

```python
# using reduce to compute maximum element from list 
print ("The maximum element of the list is : ") 
getMaxVal = functools.reduce(lambda a,b : a if a>b else b, double)
getMaxVal
```
