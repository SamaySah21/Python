LAMBDA

```python
# normal way
def add(a,b) :
    return a + b

total = add(2,3)
print(total)
```

```python
#lambda way

res = lambda a,b : a + b
total = res(2,3)
print(total)
```

FILTER

```python
# normal way

nums = [i for i in range(10)]
def getEven(n):
    return n%2 == 0

get_filtered = list(filter(getEven , nums))
print(get_filtered)
```

```python
# lambda way
get_filtered = list(filter(lambda n : n%2 ==0 , nums))
print(get_filtered)
```

MAP

```python
#normal way
def doDouble(n):
    if n in [0,2,6]:
        return n * 2

get_double = list(map(doDouble , get_filtered))
print(get_double)
```

```python
#lambda way

get_double = list(map(lambda n : n*2 if n in [0,2,6] else None, get_filtered))
print(get_double)
```

REDUCE - Apply a function cumulatively to items in an iterable to produce a single final value.

```python

from functools import reduce
```

```python
# normal way
def reducer_by_add(a,b):
    return a + b

get_reducer_value = reduce(reducer_by_add , get_filtered)
print(get_reducer_value)
```

```python
#lambda way
get_reducer_value = reduce(lambda a,b : a +b , get_filtered)

print(get_reducer_value)
```
