### Prime Number: which is divisible by 1 or itself


```python
## logic: check for all number from 2 to (number-1)
## i.e for 7 = check from 2 to 6

chk_number = 7
prime_number = False
for val in range(2, chk_number):
    if chk_number % val == 0:
        print("It is not a prime number")
        break
    else:
        prime_number = True
        pass

if prime_number == True:
    print("It is prime number")
```

```python
### for else loop : It is not in PHP , Java.. but in python there is something called for else .
chk_number = 4
for val in range(2,chk_number):
    if chk_number % val == 0:
        print("Not a prime number")
        break
else:
    print("prime Number")
```

### Fbbonacci Series: 0,1,1,2,3,5,8,13,21,...


```python
## Using List

fibbo = [0,1]
for val in range(2,10):
        temp = fibbo[val-1] + fibbo[val-2]
        fibbo.append(temp)
print(fibbo)
```

```python
## printing 

def printFibb(n):
    a = 0
    b = 1
    print(a)
    print(b)
    
    for i in range(2,n):
        c = a+b
        a = b
        b = c
        print(c)
        
printFibb(10)
```

```python
## Factorial Number : 5! = 5*4*3*2*1 = 120

def printFactorial(n):
    product = 1
    for i in range(n,1, -1):
        product = i*product
    print(product)
    
input_value = int(input("Enter number for factorial : "))

printFactorial(input_value)
```

```python
## Recursion: By default in python recursion limit is 1000.
## calling function again within function.

def printFactorialRecursion(n):
    if n == 0:
        return True

    return n * printFactorialRecursion(n-1)

input_value = int(input("Enter number for factorial : "))

a = printFactorialRecursion(input_value)
print(a)
    
```

## Frequency of word in sentence

#### Method 1:

```python
str = "hey summit hey summit hello world hello hi summit samay count word"
str_list = str.split(" ")

#print(str_list.count("hey"))

set_str = set(str_list)

for val in set_str:
    print("Count of {} is : {} ".format(val, str_list.count(val)))
```

#### Method 2:

```python
str = "hey summit hey summit hello world hello hi summit samay count word"
str_list = str.split(" ")

## creatiing unique list
uniq_str_list = []
for val in str_list:
    if val not in uniq_str_list:
        uniq_str_list.append(val)

## creating dictionary of frequency of items
dict_str = {}
for val1 in uniq_str_list:
    counter_val = 1
    for val2 in str_list:
        if val1 == val2:
            dict_str[val1] = counter_val
            counter_val =counter_val + 1

dict_str
    
```
