```python
## creating decorator for below function

def result_decor(below_result_function):
    def distinction(marks):
        for val in marks:
            if val >= 75:
                print("Distinction")
        below_result_function(marks)
    return distinction

@result_decor
def result(marks):
    for val in marks:
        if val >= 33:
            pass
        else:
            print("Fail")
            break
    else:
        print("pass")
        
result([33,56,78,89,78])

# result = result_decor(p)
```

```python
## Example 2:

## creating decorator
def printHelloDecor(below_function):
    def inner():
        print("Decorator called")
        below_function()
        print("Decorator closing")
    return inner

## base function
@printHelloDecor                       ## telling fucntion to pass as an argument in decorator.
def printHi():
    print("HI")
    
printHi()
```
