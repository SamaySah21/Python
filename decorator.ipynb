- Decorator
    - It is a function which takes another function (B) as an input and return another function (C) as an output which will be modified version of (B) basically

```python
def greet(fx):
    def mfx():
        print("Welcome !")
        fx()
        print("good bye!")
    return mfx
```

```python
def hello():
    print("hello world")

hello()
```

```python
'''whenever we wtite @greet before a function let say "hello" , 
than whenver we call hello , actually greet will be get called 
which takes this function as an input and return another function mfx 
So , to end user it looks like hello has been modified '''

@greet
def hello():
    print("hello world")

hello()
```
