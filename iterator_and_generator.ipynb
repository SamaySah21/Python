Iterator : It is an object that allows the programmer to traverse through the sequence of data without having to store the entire data in the memory . 


```python
# example LET WANT TO CREATE A list of 1000 number 
import sys
l = [x for x in range(1000)]
print(sys.getsizeof(l))

#using iterator range 
l = range(1000)
print(sys.getsizeof(l))

print("It is more memory efficient because it does not store all the values in memory at once, but generates them on the fly as needed.")
print("It is much faster than list because it does not generate all values at once, but generates them on demand.")
```

- ITERABLE
 - It is an object which one can iterate over like list , range is iterable as we can loop over it.
 - It generates an iterator when passed to iter() method.


```python
# L is an iterable
# L is not an iterator as it does not store values in memory and does not have a next() method
L = [1,2,3,4,5]
iter(L)   # will give an iterator
print(type(iter(L)))
```

```python
#how to check if an object is an iterator or not
print(hasattr(L, '__iter__'))  # True, because L is an iterable
print("It is an iterable because it has the __iter__ method, which is required for an object to be considered an iterable.")
```

```python
a = 25
print(hasattr(a, '__iter__')) # False, because a is not an iterable
print("It is not an iterable because it does not have the __iter__ method, which is required for an object to be considered an iterable.")
```
