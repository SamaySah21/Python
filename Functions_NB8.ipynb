### In python there is no concept of pass by value or pass by reference.
### value changes according to the concept whether it is mutable or not .

```python
## -------------------- integer, string is immutable  ------------------------------------

print("Address Changed for Inetger:")

def update(x):
    print("x: ", id(x) )
    x = 8
    print("x: ", id(x) )
    
a = 10
print("a: ", id(a))
update(a)

print("")
print("Address Changed for Strings:")

def update(x):
    print("x: ", id(x) )
    x = "sah"
    print("x: ", id(x) )
    
a = "samay"
print("a: ", id(a))
update(a)
```

```python
## --------------------------------  List is mutable -------------------------------------

print("Address not changed:")

def update(a_list):
    print("x: ", id(a_list) )
    a_list[2] = 90          ## changing one value of list, will not change in address(this is theidea of mutablity)

    print("x: ", id(a_list) )
    
a_list = [4,5,6]
print("a: ", id(a_list))
update(a_list)

print("")
print("")
### what if we change the whole
print("Address Changed:")

def update(a_list):
    print("x: ", id(a_list) )
    a_list = [90,2,4]          ## changing one value of list, will change in address(this is theidea of mutablity)

    print("x: ", id(a_list) )
    
a_list = [4,5,6]
print("a: ", id(a_list))
update(a_list)
```

```python
## --------------------------------- Tuple is imutable -------------------------------------

print("Id changed:")
print("")
test_tuple = (1,2,3)
print("Before", id(test_tuple))

#test_tuple[2] = 8                          ## does not support

test_tuple = (1)
print("After", id(test_tuple))
```

```python
## SET is immutable
test_set = set()
test_set = {1,2,3}
test_set[1] = 56
test_set
```

```python
## Dictionary is mutable 

print("Address not changed")
print("") 

test_dict = {1:"Samay", 2:"Sah", 3:"Flops"}
print("Before: ",id(test_dict) )

test_dict[3] = "Rocks"
print("After: ",id(test_dict) )
```

### TakeAway:
#### If mutable , address not changed like in list, dictionary.
#### If immutable , address changed like in strings, integer, tuples,set.

## TYPES OF ARGUMENT :

```python
### there are 2 types of argument:
#### Actual Parameter 
#### Formal Parameter
```

```python

def add(a,b):                       ## Formal Parameter
    c = a+b
    print(c)

add(5,6)                             ## Actual Parameter
```

#### Actual argument itself is 4 types:
##### Position, Keyword, Default, Variable Length

```python
## Position Argumnets: when argument is passed as correct position wise
def add(a,b):                       ## Formal Parameter
    c = a+b
    print(c)

add(5,6)                             ## Actual Parameter
```

```python
## Keyword Arguments: when argument is passed by using keyword of formal parameter

def add(a,b):                       ## Formal Parameter
    c = a+b
    print(c)

add(b = 5, a= 6)                             ## Actual Parameter
```

```python
## Variable Length : There are cases when we dont know how many parameters we are going to pass, the we can use variable length parameter.
## Denoted as *parametername
## Accept Value as tuple

def add(a, *b):
    
    total_tuple = a
    
    for val in b:
        total_tuple += val 
    
    print(total_tuple)
    
add(2,3,4,5,6)
```

```python
## Default : If no value passed, then value which is passed in formal paramater will be taken by default.

def add(a= 5, b= 6):
    print(a)
    print(b)

add(10)
```

```python
## args and kwargs
def fetchArguments(*args, **kwargs):
    print(args) ## return tuples of arguments
    print(kwargs) ## returns key value pair (dictionary form)
    
    ## to print key value pair 
    print("")
    for i,j in kwargs.items():
        print(i," and ",j)
fetchArguments("samay","sah","samay", "is", subject1 = "Maths" , subject2 = "Python")
```

```python
## Small Program :

def count(list):
    even =odd = 0
    for val in list:
        if val%2 == 0:
            even+= 1
        else:
            odd += 1
    return even, odd

list = [1,2,3,4,5,56,6,89,7,8,55,9,10]    
even, odd = count(list)
print(even)
print(odd)
```
