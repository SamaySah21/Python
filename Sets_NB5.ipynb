### SETS
#### Collection of unique elements
#### unordered and unidexed
#### throw away duplicate value
#### Sets are mutable

```python
### Create empty set
#  empty_set = {}           ----- dont create set like this as it will create empty dictionary
empty_set = set()

### However we can create list and tuple like this 
empty_list = []
empty_list = list()

empty_tuple = ()
empty_tuple = tuple()
```

```python
## Set has curly braces syntax
new_set = {1, 6,9,10,5,8,3}
new_set
```

##### We can see that set is unorderderd and unidexed as it ouputs randomly

```python
## Also set can be created using "set constructor". THis is also called as type casting.
set_create = set([1,5,7,0,"gfh"])
set_create
```

```python
## Unidexed : set cant fetch value by using index 
new_set[2]
```

```python
## Set is immutable
new_set[2] = "change value"
```

## Use of SETS

```python
### SETS are mostly used for this .. Actually sets are optimized for this
### Generally we define a set of elements , than we find whether this element is present in set or not
print(6 in new_set)
```

```python
## Sets are very optimized for union, intersection etc.
cs_courses = {"maths", "history", "civis", "art"}
non_cs_courses = {"commerce", "history", "finance", "economics"}
print(cs_courses.intersection(non_cs_courses))
print("")
print(cs_courses.union(non_cs_courses)) ## Add and removes duplicate values
print("")
print(cs_courses.difference(non_cs_courses))
print("")
print(non_cs_courses.difference(cs_courses))
```
