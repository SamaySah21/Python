## Dictionary :

#### Key Value Pair
#### Mutable
#### Same as Hasmap in Java / Associative Array in PHP.
#### unique value
#### unordered
#### indexed.
#### values can be change but key cannt be changed

```python
student = {"name":"samay", "age":25, "courses":["maths", "art"], 1:"one"}
print(student)
print("")
print(student["courses"])
print("")
print(student[1])
print("")
```

```python
## change the value ..dictionary is mutable
student["name"] = "samay sah"
student
```

```python
## if key is not present it gives error
student[2]
```

```python
## to avoid error use get functions
student.get(2)
```

```python
## in case key is not pressent than we can put some default value place instead
student.get(2, "no value found")
```

```python
## we can converrt 2 list into one dictinoary
key_list = [1,4,6,8,9]
val_list = ["sa", "na", "ma", "sa", "ha"]
comb_dict = dict(zip(key_list, val_list))
comb_dict

## infact a dictinalry can be a combination of simple key, values / list, dictinlary itself , inside it.
```

```python
mix_dict = {1:"abc", 2: ["a", "b", "c"], 3 : ("ab", "cd"), 4:{"a":"b", "c":3}}
mix_dict
```

```python
# change value in list
mix_dict[2][1] = "changed_val"
mix_dict
```

```python
# change value in tuple(item assignment is not supported)
mix_dict[3][1] = "cannt change as tuple is immutable"
```

```python
## can add /replace more value : dictionary is mutable
comb_dict[4] = "new"
comb_dict["onemore"] = ["onemoreval", "twomoreval", "threemoreval"]
comb_dict
```

```python
## item() : return index value of dictionry ... useful in looping thru dictionary
print(comb_dict.items())
```

```python
## Accesing dictionary index and value using for loop
for index, value in comb_dict.items():
    print("Index is {} and Value is {}". format(index, value))
```
